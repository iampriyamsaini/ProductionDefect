```python
# Step1 - Importing necessary libraries 
# Step2 - Read csv file

import pandas as pd
import numpy as np

data = pd.read_csv(r"C:\Users\priya\Downloads\Predicting Manufacturing Defects.csv")
data
```

```python
# Step3 - Check null values

data.isnull().sum()
```

```python
# Step4 - checking if there is any categorical column/feature in our dataset if present,then convert it into numerical

data.info()  #we can see here is no categorical column is present in our dataset
```

```python
# Step5 - Define input/output variables

din = data.drop(columns=['DefectStatus'])
dout = data['DefectStatus']

din.columns
```

```python
# Step 6 Data Splitting into training and testing datasets

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(din,dout,test_size=0.2,random_state=10)

print(data.shape)
x_train.shape,y_train.shape,x_test.shape,y_test.shape

```

```python
# Step7 fitting the training data into  model

from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=20)
rf_model.fit(x_train,y_train)
```

```python
x_test.head(8)
```

```python
# Step 8 Predicting the model

y_pred = rf_model.predict(x_test)
y_pred
```

```python
# Predicting a particular row 

y_pred1 = rf_model.predict(np.array([202,13175.403780,86.648534,3.121492,63.463494,9,236.439301]).reshape(1,-1)) # Reshaping the 1D array to a 2D array with 1 row and 16 columns
y_pred1
```

```python
y_pred2 = rf_model.predict(np.array([495,7143.139420,95.927814,2.195760,77.811791,7,408.406602]).reshape(1,-1)) # Reshaping the 1D array to a 2D array with 1 row and 16 columns
y_pred2
```

```python
# Step9 - Check the Accuracy of model

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

accuracy = accuracy_score(y_test,y_pred)*100
conf_matrix = confusion_matrix(y_test,y_pred)
clas_report = classification_report(y_test,y_pred)

print("Accuracy Score \n",accuracy,"\n\n")
print("Confusion Matrix \n",conf_matrix,"\n\n")
print("Classification Report \n",clas_report,"\n\n")
```

```python
# save the model 

import joblib
joblib.dump(rf_model,'Production Defect Status.pkl')
```
